```python
# Extract required data from tweets-BBC
#Pull into a DataFrame the tweet's source account, its text, its date, 
# and its compound, positive, neutral, and negative sentiment scores.
tweet_text = []
source_acct = []
create_date = []
for tweet in public_tweets_bbc["statuses"]:
    text = (tweet["text"])
    tweet_text.append(text)
    source = (tweet["id"])
    source_acct.append(source)
    date = (tweet["created_at"])
    create_date.append(date)

summary_data_bbc = {"Source Account": source_acct, "Text": tweet_text, "Date": create_date}
bbc_summary_df = pd.DataFrame(summary_data_bbc)
bbc_summary_df.head()
```

```python
# Run Vader Analysis on each tweet and find the average sentiments - BBC
compound_list = []
positive_list = []
negative_list = []
neutral_list = []
text_counter = []
counter = 0

# Loop through all tweets
for tweet in public_tweets_bbc["statuses"]:

    # Run Vader Analysis on each tweet
    compound = analyzer.polarity_scores(tweet["text"])["compound"]
    pos = analyzer.polarity_scores(tweet["text"])["pos"]
    neu = analyzer.polarity_scores(tweet["text"])["neu"]
    neg = analyzer.polarity_scores(tweet["text"])["neg"]

    # Add each value to the appropriate array
    compound_list.append(compound)
    positive_list.append(pos)
    negative_list.append(neg)
    neutral_list.append(neu)

    # Add to counter
    counter = counter + 1
    text_counter.append(counter)
    
bbc_summary2_data = {"Source Account": source_acct, "Text": tweet_text, "Date": create_date,
                "Compound Score": compound_list, "Positive Score": positive_list, "Negative Score": negative_list, 
                "Neutral Score": neutral_list, "Counter": text_counter}
bbc_summary2_df = pd.DataFrame(bbc_summary2_data)
bbc_summary2_df.tail()
```
